```python
xSize = 300
ySize = 300

@loop(600, xSize, ySize)
def rainbow_blob(screen: canvas, pen: writer, i: int):
	import colorsys
	theta = i * .01

	for j in range(20):
		phase = j * 2 * math.pi / 20
		r = 40 * (2 + math.sin(i * .04 + phase))
		(r_, g, b) = colorsys.hls_to_rgb((j / 20), .5, 1)

		pen.drawCircle(xSize // 2 + r * math.sin(theta + phase),
				ySize // 2 + r * math.cos(theta + phase),
				10,
				color()(r=r_ * 255, g=g * 255, b=b * 255, a=100 + 100 * (j / 20)), # type: ignore
				True)
```

```python
xSize = 300
ySize = 360

v_ball = 0
y_ball = 200

@loop(600, xSize, ySize)
def ball_down(screen: canvas, pen: writer, i: int):
	global v_ball, y_ball

	v_ball -= .98 # Gravitational acceleration

	y_ball += v_ball


	pen.drawCircle(xSize // 2, y_ball, 40, color()(r=255, g=60, b=60), True)
	screen.pause(16)
     
	if y_ball < 40:
		y_ball = 40
		v_ball = -v_ball * .8
		if abs(v_ball) < 1:
			return False # End the animation
```
